```python
import os
import yaml
import random
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
%matplotlib inline
```

```python
prod = '06_C300'
#prod = '02_X101 & X150'
#prod = '03_X103 & X150'
#prod = '05_FE'

path = os.path.join(prod,'train')  
print(path)

rows=4
cols=5
idx=1
axes=[]
fig = plt.figure(figsize=[10,10]) #1000*1000
fig = plt.figure(figsize=[14,7]) #1400*700


for position in os.listdir(path):
    img_path = os.path.join(path,position)
    img_name = os.listdir(img_path)[0]
    tmp = Image.open(os.path.join(img_path,img_name))
    
    axes.append(fig.add_subplot(rows,cols,idx))
    axes[-1].set_title(position)
    axes[-1].axis('off')
    plt.imshow(tmp)
    idx+=1
    
fig.tight_layout()
plt.show()
```
